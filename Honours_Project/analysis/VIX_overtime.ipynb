```python
import pandas as pd
import seaborn as sns
import math
```

```python
VIX = 18.05
def calc_month(period):
    p = math.sqrt(1/period)
    return VIX/p
    
```

```python
denom = 52
df = pd.DataFrame([i for i in range(1,denom+1)], columns=['period'])
df['denom'] = denom
df['T'] = df['period'] / denom

df['P'] = df['T'].apply(calc_month)
```

```python
sns.scatterplot(data=df, x='period', y='P')
```
